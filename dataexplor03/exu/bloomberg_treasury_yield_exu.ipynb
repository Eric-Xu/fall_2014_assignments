```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.mpl_style', 'default')

from sklearn.cross_validation import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
orig_data = pd.read_csv("/ga_fall_2014/fall_2014_assignments/dataexplor03/exu/SuperHappyFunDataSet.csv")
data = orig_data.copy()
```

### Remove NAs

```python
data = data.dropna() 
data.info()
```

### No need for so many date columns

```python
data.drop("OIS_date", axis=1, inplace=True)
data.drop("Oil_date", axis=1, inplace=True)
data.drop("US_equity_index_date", axis=1, inplace=True)
data.drop("US10Y_date", axis=1, inplace=True)
data.drop("USD_trade_weighted_date", axis=1, inplace=True)

data.info()
```

### Convert to Standardized Normal Distribution

```python
snd_data = (data - data.mean()) / data.std()
```

```python
snd_data.head()
```

### Examine correlations with Treasury Yield

```python
summary = snd_data.corrwith(snd_data.US10Y_yield)
print summary.order(ascending=False)
```

```python
columns = ["LIBOR_price", "OIS_price", "Oil_price", "USD_trade_weighted_index", "US_equity_index", "US10Y_yield"]
snd_data[columns].plot(subplots=True, figsize=(12,12))
```

### General Observations

It seems like after oil prices crashed in the end of 2008, some of the correlation patterns between the 10 year treasury yield during the previous few years became reversed. For example, the US equity index moved together with the treasury yield up until 2009, and then started diverging after that. Due to this behavior, it might be better to only look at the subset of data after 2009 when making our prediction.
