```python
x = data.rm
x.shape
```

```python
new_x = x.reshape(506,1)
new_x.shape
```

```python
y = data.medv
new_y = y.reshape(506,1)
new_y.shape
```

```python
model = LinearRegression(fit_intercept=True)
model.fit(new_x, new_y)
print ("Model coefficient: %.5f, and intercept: %.5f" % (model.coef_, model.intercept_))
```

```python
x_test = np.linspace(3,9,100).reshape(100,1)
y_hat = model.predict(x_test)
```

```python
plt.plot(x,y,'o',alpha=.5)
plt.plot(x_test,y_hat)
plt.title('Boston Housing')
plt.xlabel('rooms')
plt.ylabel('price')
```
