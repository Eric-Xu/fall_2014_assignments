```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.mpl_style', 'default')

def underline(astr): return "{}\n{}".format(astr, "-"*len(astr))
```

```python
training_data = pd.read_csv("/ga_fall_2014/fall_2014_assignments/dataexplor02/exu/kaggle_data/training.csv")
test_data = pd.read_csv("/ga_fall_2014/fall_2014_assignments/dataexplor02/exu/kaggle_data/sorted_test.csv")
print(underline("Training data info"))
print(training_data.info())
```

## Isolate and format predictor variables (Xs)

```python
soil_properties = ["Ca", "P", "pH", "SOC", "Sand"]
```

```python
X = training_data[[col for col in training_data.columns if col not in soil_properties]]
```

```python
X.head()
```

```python
X.set_index("PIDN", inplace=True)
```

```python
X.head()
```

### Replace the original 'Depth' column with two binary columns

```python
dummies = pd.get_dummies(X['Depth'], prefix='Depth')
```

```python
dummies.head()
```

```python
X[dummies.columns] = dummies
```

```python
X.head()
```

```python
X.drop('Depth', axis=1, inplace=True)
```

## Isolate response variables (ys)

```python
ys = training_data[[col for col in training_data.columns if col in ["PIDN"] + soil_properties]]
```

```python
ys.head()
```

```python
ys.set_index("PIDN", inplace=True)
```

```python
ys.head()
```

## Plot spectra graph

```python
orig_X, orig_ys = X.copy(), ys.copy()
```

```python
# if a column in the original table starts with "m", add it to the spectra_cols list; then reverse the order of the list.
spectra_cols = [col for col in orig_X.columns if col.startswith("m")][::-1]
```

```python
len(spectra_cols)
```

```python
spectra_cols[:5]
```

```python
range(10)[::-1]
```

```python
spectra_nums = [float(s[1:]) for s in spectra_cols]
```

```python
spectra_nums[:5]
```

```python
assert(spectra_nums == sorted(spectra_nums))
```

```python
yy, xx = spectra_nums, orig_X[spectra_cols]
```
